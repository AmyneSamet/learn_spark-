```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("PySparkExample").getOrCreate()
```

```python
data = [("Alice", 25), ("Bob", 30), ("Charlie", 35)]
#Create a DataFrame
data = [("Alice", 25), ("Bob", 30), ("Charlie", 35)]
df = spark.createDataFrame(data, ["Name", "Age"])
```

```python
df.show()
```

```python
df_filtered = df.filter(df["Age"] > 30)
df_filtered.show()
```

```python
df.select("Name", "Age").show()
```

```python
from pyspark.sql import functions as F

# Add a new column "Gender" with the value "Female" for all rows
df_with_gender = df.withColumn("Gender", F.lit("Female"))

# Show the result
df_with_gender.show()
```

```python
# Filtering rows based on a condition
df_filtered = df.filter(df["Age"] > 30)
df_filtered.show()
```

```python
# Sorting the DataFrame
df_sorted = df.orderBy(df["Age"])
df_sorted.show()
```

```python
from pyspark.sql import functions as F

# Grouping by 'Gender' and calculating the average 'Age'
df_grouped = df_with_gender.groupBy("Gender").agg(F.avg("Age").alias("Average_Age"))

# Show the result
df_grouped.show()
```
